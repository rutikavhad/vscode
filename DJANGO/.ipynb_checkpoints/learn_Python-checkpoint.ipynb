```python
#arithmatic operations on integers
num1=456
num2=123
#num3=123.456

print(num1+num2) #this do addition of two number = 579
print(num1/num2) #divide number =3.70...
print(num1//num2) # divide but in as integer = 3
print(num1*num2) # multipy this 
print(num1**num2) #num1 rease to num2 like squre of number num1 to num2
print(num1++num2) #same as +
print(num1-num2) # sub number
print(num1--num2) # -- become a +
print(num1---num2) # --- = -- is + and - is= -
print(num1+++num2) #same
print(num1%num2) # mod

num1+=123 # this add in number num1 456+123 only
#num1=+123 # this not add any only print 123
#num1{+,/,-,*}=123  this can  do anything
print(num1)
```
