```python
# A program to find the first number divisible by 7, 
# while skipping even numbers.

for num in range(1, 20):
    # 1. 'continue' - Skip even numbers
    if num % 2 == 0:
        continue  # Jump back to the start of the loop for the next number
    
    print(f"Checking odd number: {num}")

    # 2. 'break' - Stop entirely if we find our target
    if num % 7 == 0:
        print(f"üíé Found it! {num} is divisible by 7.")
        break  # Exit the loop immediately
else:
    # 3. 'else' - Only runs if the loop finishes WITHOUT hitting a 'break'
    print("Checked all numbers and didn't find a match.")

print("Loop process complete. ‚úÖ")
```

```python
class Car:  # Convention: Class names are usually singular and Capitalized
    def __init__(self, hp, model, brand,year):
        self.power = hp
        self.model = model
        self.brand = brand
        self.__year=year
        
    def infoo(self):
        # Use 'self' to access the object's attributes
        #year=self.__year # this access privete data (__name)
        print(f"hp={self.power} model={self.model} brand={self.brand} year={self.__year}")

car1 = Car(456, "GT", "Ford",2030)
car2 = Car(12, "Nano", "Tata",3020)

car2.infoo()
```

```python
class person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def get(self):
        print(f"name={self.name} and age = {self.age}")
        
        
class uperson(person):
    def __init__(self,name,age,dob):
        #self.name=name;
        #self.age=age   ##ue can use this a
        super().__init__(name,age) #also this can use as super() block
        self.dob=dob
        
    def get(self):
        print(f"name={self.name} and age = {self.age} and DOB is{self.dob}")
        
        
person1=person("king",24)
person2=uperson("raja",23,10)

person1.get()
person2.get()
```
