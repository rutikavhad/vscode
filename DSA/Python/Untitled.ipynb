```python
class Node:
    def __init__(self,value):
        self.value=value
        self.next=None


n=int(input("enter how many linked  list you want "))
#fhead=int(input("enter 1st number"))
head=None
i=0
while i<n:
    i+=1
    new_node=int(input(f"enter{i}st number"))
    new_nod=Node(new_node) #Node Object not use direct node not use number
    if head is None:
        head=new_nod #1st element is head
    else:
        current=head #make chain do reduce complex
        while current.next is not None:
            current=current.next #if next will not empty then next node is head(currnet)
        current.next=new_nod


#print(f"added{new_node}")

#print all
# temp=head 
# while temp.next is not None:   # this not print all -1 last miss
#     print(temp.value)
#     temp=temp.next

temp=head
while temp: # this can print total
    print(temp.value)
    temp=temp.next
```
