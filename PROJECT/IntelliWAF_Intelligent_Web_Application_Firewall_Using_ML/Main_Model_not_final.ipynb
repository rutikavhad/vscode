```python
import pandas as pd
import re
import urllib.parse

# -----------------------------
# Load CSV safely
# -----------------------------
df = pd.read_csv(
    "intelliwaf_combined_new.csv",
    low_memory=False
)

# -----------------------------
# Pattern Libraries
# -----------------------------

SQLI_PATTERNS = [
    r"union\s+select",
    r"select.+from",
    r"insert\s+into",
    r"update\s+\w+\s+set",
    r"delete\s+from",
    r"drop\s+table",
    r"or\s+1\s*=\s*1",
    r"or\s+'1'\s*=\s*'1'",
    r"'--",
    r"--\s",
    r"/\*.*\*/",
    r"sleep\s*\(",
    r"benchmark\s*\(",
    r"load_file\s*\(",
    r"information_schema",
    r"@@version",
    r"xp_cmdshell",
]

XSS_PATTERNS = [
    r"<\s*script",
    r"</\s*script",
    r"javascript:",
    r"onerror\s*=",
    r"onload\s*=",
    r"onclick\s*=",
    r"onmouseover\s*=",
    r"<\s*img",
    r"<\s*svg",
    r"<\s*iframe",
    r"alert\s*\(",
    r"document\.cookie",
    r"%3cscript",
    r"%3cimg",
]

CMDI_PATTERNS = [
    r";\s*\w+",
    r"\|\|",
    r"\&\&",
    r"`.+?`",
    r"\$\(.+?\)",
    r"\bwhoami\b",
    r"\bid\b",
    r"\buname\b",
    r"\bls\b",
    r"\bpwd\b",
    r"\bcat\b",
    r"\bifconfig\b",
    r"\bip\s+a\b",
    r"\bwget\b",
    r"\bcurl\b",
    r"%0a",
    r"%0d",
]

TRAVERSAL_PATTERNS = [
    r"\.\./",
    r"\.\.\\",
    r"%2e%2e",
    r"%2f",
    r"/etc/passwd",
    r"/etc/shadow",
    r"/proc/self/environ",
    r"win\.ini",
    r"boot\.ini",
    r"php://",
    r"file://",
]

UPLOAD_PATTERNS = [
    r"\.php$",
    r"\.php\.",
    r"\.phtml",
    r"\.php5",
    r"\.jsp",
    r"\.asp",
    r"\.exe",
    r"multipart/form-data",
    r"content-disposition",
]

# -----------------------------
# Helper Function
# -----------------------------

def match_any(patterns, text):
    for p in patterns:
        if re.search(p, text):
            return True
    return False

# -----------------------------
# Labeling Function
# -----------------------------

def label_row(row):
    raw_text = f"{row.get('url','')} {row.get('path','')} {row.get('body','')}"
    decoded = urllib.parse.unquote_plus(str(raw_text)).lower()

    # ---- Priority order matters ----
    if match_any(SQLI_PATTERNS, decoded):
        return 1  # SQL Injection

    if match_any(XSS_PATTERNS, decoded):
        return 2  # XSS

    if match_any(CMDI_PATTERNS, decoded):
        return 3  # Command Injection

    if match_any(TRAVERSAL_PATTERNS, decoded):
        return 4  # Directory Traversal / LFI

    if row.get("has_file_upload", 0) == 1 or match_any(UPLOAD_PATTERNS, decoded):
        return 5  # File Upload

    return 0  # Normal

# -----------------------------
# Apply Labels
# -----------------------------

df["label"] = df.apply(label_row, axis=1)

# -----------------------------
# Save Output
# -----------------------------

df.to_csv("labeled_attacks.csv", index=False)

print("Label distribution:")
print(df["label"].value_counts())
```

```python
import re
import urllib.parse

SQLI_REGEX = re.compile(
    r"(union\s+select|select.+from|insert\s+into|update\s+\w+\s+set|"
    r"delete\s+from|drop\s+table|or\s+1\s*=\s*1|sleep\s*\(|benchmark\s*\(|"
    r"load_file\s*\(|information_schema|@@version|xp_cmdshell|'--|--\s)",
    re.IGNORECASE
)


def is_sqli_content(row):
    text = f"{row.get('url','')} {row.get('path','')} {row.get('body','')}"
    decoded = urllib.parse.unquote_plus(str(text))
    return bool(SQLI_REGEX.search(decoded))

sqli_by_content = df[df.apply(is_sqli_content, axis=1)]

print(len(sqli_by_content))
```

```python
sqli_strict = df[
    (df["label"] == 1) &
    (df.apply(is_sqli_content, axis=1))
]

print(len(sqli_strict))
```

```python
SQLI_PAYLOADS = [
    # UNION
    "union select", "union all select", "union distinct select",
    "union select null", "union select 1", "union select 1,2",
    "union select username", "union select password",
    "union select table_name", "union select column_name",
    "union select information_schema",

    # BOOLEAN
    "or 1=1", "or '1'='1'", "or \"1\"=\"1\"",
    "and 1=1", "and 1=0",
    "or true", "and false",
    "or exists", "and exists",
    "or not exists",

    # COMMENTS
    "--", "#", "/*", "*/", ";--", ";#",

    # TIME BASED
    "sleep(", "benchmark(", "pg_sleep(",
    "waitfor delay",

    # ERROR BASED
    "sql syntax", "mysql_fetch", "mysql_num_rows",
    "warning: mysql", "odbc_exec", "pg_query",
    "quoted string not properly terminated",
    "unclosed quotation",

    # ENUMERATION
    "information_schema", "table_schema", "table_name",
    "column_name", "@@version", "version()",
    "database()", "current_user", "system_user",

    # OBFUSCATION
    "0x", "char(", "concat(", "hex(", "unhex(",
]
```

```python
import re

SQLI_REGEX_LIST = [
    re.compile(re.escape(p), re.IGNORECASE)
    for p in SQLI_PAYLOADS
]
```

```python
import urllib.parse

def is_sqli_safe(row):
    text = f"{row.get('url','')} {row.get('path','')} {row.get('body','')}"
    decoded = urllib.parse.unquote_plus(str(text)).lower()

    for rx in SQLI_REGEX_LIST:
        if rx.search(decoded):
            return True
    return False
```

```python
missed_sqli = df[
    (df["label"] != 1) &
    (df.apply(is_sqli_safe, axis=1))
]

print("Missed SQLi samples:", len(missed_sqli))
```

```python
import pandas as pd

df = pd.read_csv("labeled_attacks.csv", low_memory=False)

# consider only "Normal" traffic
normal_df = df[df["label"] == 0]

# login-like paths
login_mask = normal_df["path"].str.contains(
    "login|signin|admin|auth",
    case=False,
    na=False
)

bf_candidates = (
    normal_df[login_mask]
    .groupby("client_ip")
    .size()
    .sort_values(ascending=False)
)

bf_candidates.head(10)
```

```python
# convert timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

normal_df = df[df["label"] == 0]

# requests per IP per second
dos_rate = (
    normal_df
    .groupby([
        "client_ip",
        normal_df["timestamp"].dt.floor("s")
    ])
    .size()
    .reset_index(name="req_per_sec")
)

dos_rate.sort_values("req_per_sec", ascending=False).head(10)
```

```python
dos_paths = (
    normal_df
    .groupby(["client_ip", "path"])
    .size()
    .sort_values(ascending=False)
)

dos_paths.head(10)
```
