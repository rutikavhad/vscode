```python
import csv
from datetime import datetime, timedelta
import random

file_path = "offline_attack_dataset_200.csv"

header = [
    "timestamp","client_ip","method","url","path","body","body_len",
    "body_entropy","special_char_count","path_depth","query_length",
    "has_file_upload","response_status","response_length"
]

xss_payloads = [
    "<script>alert(1)</script>",
    "%3Cscript%3Ealert(1)%3C/script%3E",
    "<img src=x onerror=alert(1)>",
    "<svg onload=alert(1)>",
    "<body onload=alert(1)>"
]

sqli_payloads = [
    "' OR '1'='1",
    "' UNION SELECT null--",
    "admin'--",
    "1;DROP TABLE users"
]

cmd_payloads = [
    "ls;",
    "cat /etc/passwd",
    "whoami",
    "id",
    "ping -c 4 127.0.0.1"
]

trav_payloads = [
    "../../../../etc/passwd",
    "..%2F..%2F..%2Fetc%2Fpasswd",
    "/etc/shadow"
]

upload_payloads = [
    "filename=shell.php&content=<?php system($_GET['c']); ?>",
    "filename=test.jsp&content=<% Runtime.getRuntime().exec(request.getParameter(\"c\")); %>"
]

rows = []
start_time = datetime(2025,1,1,10,0,0)

for i in range(2500):
    ts = (start_time + timedelta(minutes=i)).isoformat() + "Z"
    attack_type = random.choice(["xss","sqli","cmd","trav","upload"])

    if attack_type == "xss":
        body = random.choice(xss_payloads)
        path = "/xss"
        method = "GET"
        has_upload = 0
    elif attack_type == "sqli":
        body = random.choice(sqli_payloads)
        path = "/search"
        method = "GET"
        has_upload = 0
    elif attack_type == "cmd":
        body = random.choice(cmd_payloads)
        path = "/cmd"
        method = "GET"
        has_upload = 0
    elif attack_type == "trav":
        body = random.choice(trav_payloads)
        path = "/file"
        method = "GET"
        has_upload = 0
    else:
        body = random.choice(upload_payloads)
        path = "/upload"
        method = "POST"
        has_upload = 1

    url = f"http://test{path}?q={body}"
    rows.append([
        ts,"offline",method,url,path,body,len(body),
        round(random.uniform(3.5,6.5),2),
        sum(1 for c in body if not c.isalnum()),
        path.count("/"),
        len(body),has_upload,
        random.choice([200,302,403,500]),
        random.randint(200,1600)
    ])

with open(file_path,"w",newline="",encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)

file_path
```

```python
import csv
from datetime import datetime, timedelta
import random

file_path = "dvwa_benign_normal_users_500.csv"

header = [
    "timestamp","client_ip","method","url","path","body",
    "body_len","body_entropy","special_char_count",
    "path_depth","query_length","has_file_upload",
    "response_status","response_length"
]

start_time = datetime(2025, 12, 24, 7, 30, 0)

normal_sql = [
    "id=1",
    "id=2",
    "user=admin",
    "search=test",
    "search=hello"
]

normal_cmd = [
    "ip=127.0.0.1",
    "ip=localhost",
    "ip=192.168.1.1"
]

normal_upload = [
    "filename=image.jpg",
    "filename=photo.png",
    "filename=resume.pdf",
    "filename=document.docx"
]

normal_xss = [
    "name=John",
    "comment=Hello world",
    "msg=Nice website"
]

normal_fi = [
    "page=include.php",
    "page=home.php",
    "page=about.php"
]

rows = []

for i in range(20000):
    ts = (start_time + timedelta(seconds=i * 4)).isoformat() + "Z"
    client = f"('127.0.0.1', {random.randint(30000,40000)})"
    choice = random.choice(["sql","cmd","upload","xss","fi"])

    if choice == "sql":
        body = random.choice(normal_sql)
        path = "/dvwa/vulnerabilities/sqli/"
        method = "GET"
        has_upload = 0
        url = f"http://192.168.122.170{path}?{body}"
    elif choice == "cmd":
        body = random.choice(normal_cmd)
        path = "/dvwa/vulnerabilities/exec/"
        method = "POST"
        has_upload = 0
        url = f"http://192.168.122.170{path}"
    elif choice == "upload":
        body = random.choice(normal_upload)
        path = "/dvwa/vulnerabilities/upload/"
        method = "POST"
        has_upload = 1
        url = f"http://192.168.122.170{path}"
    elif choice == "xss":
        body = random.choice(normal_xss)
        path = "/dvwa/vulnerabilities/xss_r/"
        method = "GET"
        has_upload = 0
        url = f"http://192.168.122.170{path}?{body}"
    else:
        body = random.choice(normal_fi)
        path = "/dvwa/vulnerabilities/fi/"
        method = "GET"
        has_upload = 0
        url = f"http://192.168.122.170{path}?{body}"

    rows.append([
        ts,
        client,
        method,
        url,
        path,
        body,
        len(body),
        round(random.uniform(1.2, 2.8), 3),
        sum(1 for c in body if not c.isalnum()),
        path.count("/"),
        len(body),
        has_upload,
        200,
        random.randint(2500, 4800)
    ])

with open(file_path, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(rows)

file_path
```
