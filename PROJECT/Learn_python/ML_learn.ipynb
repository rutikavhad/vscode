```python
import numpy as np
import pandas as pd
from scipy import stats

def mean(x):
    return sum(x) / len(x)

def variance(x):
    m = mean(x)
    return sum((i - m) ** 2 for i in x) / len(x)

data = np.array([2,4,6,8,10])
print(mean(data), variance(data))
```

```python
def gradient_descent(X, y, lr=0.01, epochs=1000):
    m, b = 0.0, 0.0
    n = len(X)

    for _ in range(epochs):
        y_pred = m*X + b
        dm = (-2/n) * sum(X * (y - y_pred))
        db = (-2/n) * sum(y - y_pred)
        m -= lr * dm
        b -= lr * db

    return m, b
```
