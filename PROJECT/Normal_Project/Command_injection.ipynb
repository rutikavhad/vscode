```python
strin={
        "2025-01-01T22:20:00.000000Z,offline,GET,\"http://test/fi?page=data://text/plain,<?php phpinfo();?>\",/fi,\"data://text/plain,<?php phpinfo();?>\",COMMAND_INJECTION|PATH_TRAVERSAL,36.0,4.92,13.0,False,[0 0 1 1 0 0 0]"
    }
```

```python

# The dangerous characters
cmd_signs = [";", "&&", "||", "|", "`", "$("]

# Loop through each string in your list
for text in strin:
    print(f"Checking: {text}")
    
    # Check if any dangerous sign is INSIDE the text
    found = False
    for sign in cmd_signs:
        if sign in text:
            print(f"  ⚠️ ALERT: Command Injection sign '{sign}' found!")
            found = True
            
    if not found:
        print("  ✅ This input is clean.")
```
