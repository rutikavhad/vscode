```python
import csv
import random
import datetime

# Output file
OUTPUT = "synthetic_captures.csv"

# Safe attack placeholders (you will replace later on your machine)
ATTACK_PLACEHOLDERS = {
    "SQLi": "<SQL_INJECTION_PAYLOAD>",
    "XSS": "<XSS_PAYLOAD>",
    "LFI": "<LOCAL_FILE_INCLUSION_PAYLOAD>",
    "CMD": "<COMMAND_INJECTION_PAYLOAD>",
    "NORMAL": "<NORMAL_REQUEST_BODY>"
}

METHODS = ["GET", "POST"]
USER_AGENTS = [
    "Mozilla/5.0 (X11; Linux x86_64; rv:128.0) Gecko Firefox",
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
]

URLS = [
    "/login.php",
    "/search.php",
    "/profile.php",
    "/products.php",
    "/comments.php"
]

HOST = "192.168.1.10"

def random_timestamp():
    return datetime.datetime.utcnow().isoformat() + "Z"

def generate_capture_block(event_type):
    timestamp_now = random_timestamp()
    client_ip = f"127.0.0.1:{random.randint(30000, 60000)}"
    method = random.choice(METHODS)
    path = random.choice(URLS)
    url = f"http://{HOST}{path}"
    user_agent = random.choice(USER_AGENTS)

    body = ATTACK_PLACEHOLDERS[event_type]

    raw_text = f"""---- CAPTURE START ----
timestamp: {timestamp_now}
client_ip: {client_ip}
method: {method}
url: {url}
path: {path}
---- HEADERS ----
Host: {HOST}
User-Agent: {user_agent}
Accept: */*
---- QUERY ----

---- BODY ----
{body}
---- CACHE-RELATED HEADERS ----
Cache-Control: (none)
Pragma: (none)
Expires: (none)
If-Modified-Since: (none)
If-None-Match: (none)
---- CAPTURE END ----
"""

    return timestamp_now, raw_text

def generate_dataset(rows=400):
    events = ["NORMAL", "SQLi", "XSS", "LFI", "CMD"]

    with open(OUTPUT, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["timestamp", "raw_text"])

        for _ in range(rows):
            event_type = random.choice(events)
            ts, block = generate_capture_block(event_type)
            writer.writerow([ts, block])

    print(f"Dataset created: {OUTPUT}")

# Generate 400+ rows
generate_dataset(rows=450)
```
