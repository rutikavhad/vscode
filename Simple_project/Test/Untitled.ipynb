```python
import bcrypt

password = "A9$kQz!R2@"

# bcrypt requires bytes
password_bytes = password.encode("utf-8")
print(password_bytes)

# generate salt with cost factor
salt = bcrypt.gensalt(rounds=11)
print(salt)

# hash password
hashed = bcrypt.hashpw(password_bytes, salt)
print(hashed)

# store this string in DB
stored_hash = hashed.decode("utf-8")

print(stored_hash)
```
